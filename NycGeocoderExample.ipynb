Developer Portal (http://developer.cityofnewyork.us/api/geoclient-api), and make sure you check off access to the Geoclient API for the application. Take note of the Application's ID and key. 

There is also a Geosupport Desktop Edition, which can be downloaded at:
http://www.nyc.gov/html/dcp/html/bytes/gdeguide.shtml
(only runs on Windows)

```python
# read in PLUTO dataframe as an example
import pandas as pd
df=pd.read_csv("../pluto/nyc_pluto_16v1/MN.csv")
df.columns
```

```python
df=df[['BBL','Address','BoroCode']]
print df.shape
df.head()
```

```python
df['Boro']='0'
df.loc[df['BoroCode']==1,['Boro']]='Manhattan'
df.loc[df['BoroCode']==2,['Boro']]='Bronx'
df.loc[df['BoroCode']==3,['Boro']]='Brooklyn'
df.loc[df['BoroCode']==4,['Boro']]='Queens'
df.loc[df['BoroCode']==5,['Boro']]='Staten Island'
df.head()
```

```python
df['number']=df['Address'].str.extract('(^[0-9|-]*)')
#extracting the number part of the address
df['street']=df['Address'].str.extract('(\s.+$)')
#extracting the rest of the address after the first space to the end.
df.head()
```

```python
import sys

# define a function to be applied to data frame, taking the street number,
# street name, and boro 
# and returning the bbl
def getBBL(frame):
    try:
        x = g.address(frame['number'],frame['street'],frame['Boro'])['bbl']
    except:
        e = sys.exc_info()[0]
        x = ( "Error: %s" % e )
    return x

#initialize a column, which will hold the geocoded BBL
df['geocodedBBL']=0
```

```python
#run this on a small subset of df

smalldf=df[:10]

%timeit \
smalldf['geocodedBBL'] = smalldf[['number','street','Boro']].apply(getBBL,axis=1)
smalldf
```
